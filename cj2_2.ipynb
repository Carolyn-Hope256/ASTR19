```python
import matplotlib.pyplot as plt
import numpy as np
import math
```

```python
def s(x):
    return(np.sin(x))
```

```python
def c(x):
    return(np.cos(x))
```

```python
x = np.linspace(0, 1, 100)
```

```python
figure, fig = plt.subplots(1,2)
fig[0].plot(x,s(x))
fig[0].set_title("Sin(x)")
fig[1].plot(x,c(x))
fig[1].set_title("Cos(x)")
plt.savefig("cj2_2_fig.pdf", format='pdf', bbox_inches = 'tight')
```
